```python
#slicing [a:b] (a is excluded and b is included)
list1 = ["abc",34,True,40,"male",40.5,3+2j]
print(list1[2:5])
print(list1[:5]) #from 1st data to the 5th data
```

```python
#slicing using negative index
list1 = ["abc",34,True,40,"male",40.5,3+2j]
print(list1[-3:])
```

```python
#change the item value
list1 = ["abc",34,True,40,"male",40.5,3+2j]
list1[2] = "water"
print(list1)
```

```python
#append items(adds any data to the last of the list)
list1 = ["abc",34,True,40,"male",40.5,3+2j]
list1.append("water")
list1
```

```python
#insert (insert with the location of index)
list1 = ["abc",34,True,40,"male",40.5,3+2j]
list1.insert(3,"water")
list1
```

```python
#extend (adding two oe mre items to the list)
list1 = ["abc",34,True,40,"male",40.5,3+2j]
namelist = ["Raj","Rahul","Sunil"]
list1.extend(namelist)
list1
```

```python
'''1. marks of 5 students in a list'''
marks = [78,85,90,66,88]
"Maxium marks :"
"Minimum marks :"
"total marks :"
"Average marks :"
"sort in assecing order"
```

```python
marks = [78,85,90,66,88]
print("Maxium marks :",max(marks))
print("Minimum marks :",min(marks))
print("total marks :",sum(marks))
print("Average marks :",sum(marks)/len(marks))
marks.sort()
print(marks)
```

```python
'''4. Temperature Analysis
Daily temperature readings of a city are stored in a list for 10 days.print
tasks:
Display temperatures of first an last 5 days
find highest and lowest temperature
claculate average temperature
```

```python
temp = [45,56,67,78,89,23,43,65,32,46] #in degree centigrate
print ("the temperatures of first 5 days are ",temp[:5])
print ("the temperatures of last 5 days are ",temp[-5:])
print ("the highest temperature is ",max(temp))
print ("the lowest temperature is ",min(temp))
print ("the average temperature is ",sum(temp)/len(temp))
```
